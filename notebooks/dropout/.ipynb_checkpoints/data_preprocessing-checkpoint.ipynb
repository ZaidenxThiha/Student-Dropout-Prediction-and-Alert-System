### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

### Load Datasets

```python
# 1. LOAD DATA

df_info = pd.read_csv("..\\data\\oula\\studentInfo.csv")
df_vle= pd.read_csv("..\\data\\oula\\studentVle.csv")
df_reg = pd.read_csv("..\\data\\oula\\studentRegistration.csv")
```

### Feature Engineering

```python
# 2. FEATURE ENGINEERING 
# A. Aggregate VLE Clicks (Behavioral Red Flag)
engagement = df_vle.groupby(['id_student', 'code_module', 'code_presentation'])['sum_click'].sum().reset_index()

# B. Registration Timing (Administrative Red Flag)
registration_timing = df_reg[['id_student', 'code_module', 'code_presentation', 'date_registration']]

# Merge these new features into the main student info table
df_oula = pd.merge(df_info, engagement, on=['id_student', 'code_module', 'code_presentation'], how='left')
df_oula = pd.merge(df_oula, registration_timing, on=['id_student', 'code_module', 'code_presentation'], how='left')
```

```python
print(f"Dataset shape after merging features: {df_oula.shape}")
display(df_oula.head())
```

### Data Cleaning

```python
# 3. HANDLING MISSING VALUES & DUPLICATES
df_oula = df_oula.drop_duplicates()

# Fill missing clicks with 0 (meaning they never logged in)
df_oula['sum_click'] = df_oula['sum_click'].fillna(0)

# Fill missing registration dates with the median
df_oula['date_registration'] = pd.to_numeric(df_oula['date_registration'], errors='coerce')
# Fill missing demographic text data with the most frequent value (mode)
df_oula['imd_band'] = df_oula['imd_band'].fillna(df_oula['imd_band'].mode()[0])
```

```python
# 4. DEFINING THE TARGET (Dropout Prevention Goal)
# 1 = Dropout ('Withdrawn'), 0 = Retained ('Pass', 'Fail', 'Distinction')
df_oula['is_dropout'] = df_oula['final_result'].apply(lambda x: 1 if x == 'Withdrawn' else 0)
df_oula = df_oula.drop('final_result', axis=1)
```

```python
# 5. VISUALIZATION
plt.figure(figsize=(8, 5))
sns.boxplot(x='is_dropout', y='sum_click', data=df_oula)
plt.title('Total VLE Clicks: Retained (0) vs. Dropout (1)')
plt.ylim(0, 5000) # Limit y-axis to see the boxplot clearly
plt.show()

# A. Correlation Heatmap: Identifying Important Features
# We select the key numerical risk factors to see how they correlate with 'is_dropout'
plt.figure(figsize=(8, 5))
numeric_for_corr = df_oula[['num_of_prev_attempts', 'studied_credits', 'sum_click', 'date_registration', 'is_dropout']]
corr_matrix = numeric_for_corr.corr()

# Plotting the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='RdYlBu_r', fmt=".2f", vmin=-1, vmax=1, 
            linewidths=0.5, annot_kws={"size": 12})
plt.title('Correlation Heatmap: Risk Factors vs. Dropout (OULA)', fontsize=14)
plt.show()

# B. Relationship: Previous Attempts vs. Dropout Rate
# Does failing in the past mean a student is more likely to drop out now?
plt.figure(figsize=(8, 5))
# A bar plot automatically calculates the mean (dropout rate) for each category
sns.barplot(x='num_of_prev_attempts', y='is_dropout', data=df_oula, errorbar=None, palette='Reds')
plt.title('Relationship: Previous Failed Attempts vs. Dropout Probability', fontsize=14)
plt.ylabel('Probability of Dropping Out (0.0 to 1.0)')
plt.xlabel('Number of Previous Attempts')
plt.show()

# C. Relationship: Credits Studied vs. Dropout Status
# Are students taking too many credits dropping out more?
plt.figure(figsize=(8, 5))
sns.boxplot(x='is_dropout', y='studied_credits', data=df_oula, palette='Set2')
plt.title('Relationship: Course Load (Credits) vs. Dropout Status', fontsize=14)
plt.xticks([0, 1], ['Retained (0)', 'Dropout (1)'])
plt.ylabel('Total Studied Credits')
plt.show()
```

### Data Preprocessing

```python
# 6. ENCODING CATEGORICAL VARIABLES
categorical_cols = ['code_module', 'code_presentation', 'gender', 'region', 
                    'highest_education', 'imd_band', 'age_band', 'disability']

encoder = LabelEncoder()
for col in categorical_cols:
    df_oula[col] = encoder.fit_transform(df_oula[col].astype(str))
```

```python
# 7. NORMALIZING NUMERIC FEATURES
scaler = StandardScaler()
numeric_cols = ['num_of_prev_attempts', 'studied_credits', 'sum_click', 'date_registration']

df_oula[numeric_cols] = scaler.fit_transform(df_oula[numeric_cols])

# Drop IDs as they hold no predictive value for the AI
df_preprocessed = df_oula.copy()
```

```python
# 8. EXPORT TO PROCESSED FOLDER
output_path = '..\\data\\processed\\oula_preprocessed.csv'
df_preprocessed.to_csv(output_path, index=False)

print(f"Fully processed dataset saved to: {output_path}")
print(f"Final Dataset Shape: {df_preprocessed.shape}")
```
