```python
import pandas as pd
import joblib
```

```python
# 1. Load Model, Test Data, and Student IDs
model = joblib.load('../../models/dropout/rf_dropout_model.joblib')
students_df = pd.read_csv('../../data/processed/dropout/X_test.csv')
student_ids = pd.read_csv('../../data/processed/dropout/student_ids.csv').squeeze()

# 2. Calculate Risk
risk_probabilities = model.predict_proba(students_df)[:, 1]
predictions = (risk_probabilities >= 0.40).astype(int)

# 3. Build the Clean Human-Readable Report
report_df = pd.DataFrame()
report_df['Student_ID'] = student_ids
report_df['Prediction_Band'] = pd.Series(predictions).map({1: 'band_1', 0: 'band_2'})

report_df['Numeric_Risk'] = risk_probabilities 

# 4. GENERATE THE "REASONS" FOR RISK 
def get_risk_reasons(row):
    reasons = []
    if row['sum_click'] < -0.2: reasons.append("Low Online Engagement")
    if row['date_registration'] > 0.1: reasons.append("Late Registration")
    if row['studied_credits'] > 0.5: reasons.append("High Course Load")
    if row['num_of_prev_attempts'] > 0: reasons.append("Previous Failures")
    if len(reasons) == 0: return "General Risk Profile"
    return " | ".join(reasons)

report_df['Primary_Risk_Factors'] = students_df.apply(get_risk_reasons, axis=1)

# 5. Add Intervention Status
report_df['Intervention_Status'] = pd.Series(predictions).map({
    1: 'ACTION REQUIRED: High Risk', 
    0: 'Safe: On Track'
})
report_df.loc[report_df['Intervention_Status'] == 'Safe: On Track', 'Primary_Risk_Factors'] = "N/A"

report_df = report_df.sort_values(by='Numeric_Risk', ascending=False).reset_index(drop=True)

# 7. FORMATTING: Now it is safe to add the percentage signs
report_df['Risk_Probability'] = (report_df['Numeric_Risk'] * 100).round(1).astype(str) + '%'

# Drop the hidden numeric column to keep the CSV clean
report_df = report_df.drop('Numeric_Risk', axis=1)

# 8. EXPORT TO CSV
output_path = '../../data/processed/dropout/actionable_weekly_risk_report.csv'
report_df.to_csv(output_path, index=False)

print(f"Report saved to: {output_path}")

print("\nTop 15 At-Risk Students:")
display(report_df[['Student_ID', 'Prediction_Band', 'Risk_Probability', 'Primary_Risk_Factors']].head(15))
print("\nSafest 5 Students (Bottom of the list):")
display(report_df[['Student_ID', 'Prediction_Band', 'Risk_Probability', 'Primary_Risk_Factors']].tail(10))
```
